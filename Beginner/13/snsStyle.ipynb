```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style("darkgrid")
x = np.arange(1, 11)
y1 = np.random.randint(10, 100, 10)

# Create a figure with a single subplot
plt.figure(figsize=(8, 5))

# Line Plot with Custom Aesthetics
plt.plot(x, y1, marker='D', linestyle='--', color='purple', linewidth=2, label="Custom Line")

# Customisation
plt.title("Customised Line Plot", fontsize=14, fontweight='bold')
plt.xlabel("X-Axis Label", fontsize=12)
plt.ylabel("Y-Axis Label", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()
```

Customisation Highlights:

 Seaborn Styling (sns.set_style("darkgrid"))
 Dashed Line Style (--)
 Diamond Markers (marker='D')
 Bold Title (fontweight='bold')
 Custom Legend & Grid
