```python
import numpy as np
import matplotlib.pyplot as plt


# Create sample data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # y is the sine of x

# Create the line chart
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Chart')

# Show the plot
plt.show()
```
