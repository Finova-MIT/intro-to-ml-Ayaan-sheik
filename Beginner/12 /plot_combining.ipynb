```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data
np.random.seed(42)
x = np.arange(1, 11)
y1 = np.random.randint(10, 100, 10)  # Line plot data
y2 = np.random.randint(5, 50, 10)   # Bar plot data
y3 = np.random.randint(20, 80, 10)  # Scatter plot data

```

```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# **1. Line Plot**
axes[0].plot(x, y1, marker='o', linestyle='-', color='b', label="Line Data")
axes[0].set_title("Line Plot")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].legend()

# **2. Bar Plot**
axes[1].bar(x, y2, color='g', label="Bar Data")
axes[1].set_title("Bar Chart")
axes[1].set_xlabel("Categories")
axes[1].set_ylabel("Values")
axes[1].legend()

# **3. Scatter Plot**
axes[2].scatter(x, y3, color='r', marker='s', label="Scatter Data")
axes[2].set_title("Scatter Plot")
axes[2].set_xlabel("X-axis")
axes[2].set_ylabel("Y-axis")
axes[2].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

Explanation:

    plt.subplots(1, 3, figsize=(15, 5))  Creates 3 subplots in one row.

    axes[0].plot()  Line plot with markers and labels.

    axes[1].bar()  Bar chart with custom colours.

    axes[2].scatter()  Scatter plot with square markers
