```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



dates = pd.date_range(start="2023-01-01", periods=100, freq="D")
time_series_data = np.cumsum(np.random.randn(100))  
df = pd.DataFrame({"Date": dates, "Value": time_series_data})

# Plot Time Series
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Value"], label="Time Series", color="green")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Data")
plt.legend()
plt.show()
```
