```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("matches.csv")
df['toss_and_match_winner'] = df['toss_winner'] == df['winner']

heatmap_data = pd.crosstab(df['toss_winner'], df['winner'], normalize='index') * 100  

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5)


plt.title('Heatmap: Relationship Between Toss Winner and Match Winner')
plt.xlabel('Match Winner')
plt.ylabel('Toss Winner')

plt.show()
```
